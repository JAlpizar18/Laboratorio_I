# Metodología


Para el cálculo del SNR (signal-to-noise ratio) se genera un tono mediante el lenguaje de programación Python de una frecuencia de 1 kHz el cual se usa como referencia, luego este sonido se reproduce en un parlante el cual se graba utilizando el micrófono de un celular y el micrófono de una laptop. Una vez obtenidas las grabaciones y utilizando un tono de ruido ambiente se determinan las potencias a través del uso de la densidad espectral de potencia mediante el método de Welch en Matlab. Este método en lugar de calcular el espectro de potencia directamente utilizando una transformada de Fourier, divide la señal en segmentos más cortos y calcula el espectro de potencia de cada segmento por separado. Luego, combina estos espectros de potencia segmentados para obtener una estimación más precisa del espectro total, ya que la señal varía en el tiempo. Una vez calculados estos datos se obtiene la gráfica del PSD (densidad espectral de potencia) para ambas grabaciones.  Ya con estos datos obtenidos, se aplica la ecuación para calcular el valor del SNR en el mismo programa (Matlab) y con ello se obtienen las gráficas del ruido. 

SNR (dB) = 10 ∗ log10(Potencia de la señal / Potencia del ruido) ) 

  

Para realizar el método de estimación de respuesta en frecuencia primero se debe de generar un sonido en Python de un barrido de frecuencia que va desde 50 Hz hasta 5 kHz, luego este se reproduce mediante un parlante para posteriormente grabarlo con ambos micrófonos (celular y laptop) igual que en el caso del SNR. Luego, mediante Matlab se utiliza la función fft (transformada rápida de Fourier) para calcular el análisis de frecuencia a los barridos de frecuencia tanto para el micrófono de la laptop como para el del celular y de esta manera obtener la respuesta en frecuencia y su respectiva gráfica.

   Con estos datos anteriores podemos hacer un analisis que resulta de suma importancia para observar si nuestro SNR es un aproximacion realista o no, y esto lo hacemos determinando el tipo de ruido que tenemos presente en los microfonos para lo cual tenemos que analizar la siguiente figura.

    Figura 4. Espectrograma del ruido.  

En ella podemos observar como se comporta el ruido en ambos micrófonos, se observa que para el ruido desde el celular tenemos unas líneas verticales que siguen cierto patrón, pero que son muy pequeñas, este tipo de líneas nos indica un ruido por cuantización en contraposición tenemos las mismas líneas en los primeros segundos del ruido desde la laptop, donde se observa que son mas grandes en términos de frecuencia, pero que a partir del segundo 3.5 desaparecen del resto del espectro. 

Es importante recalcar  que la tasa de muestreo es diferente en ambos micrófonos, para el celular tenemos una tasa de 48Khz y para la laptop una tasa mas normalizada de 44.1 Khz, es probable que debido a esta diferencia y a componentes físicos del diseño de los micrófonos podemos ver esta diferencia solo en los primeros 3.5 segundos del espectrograma de la laptop. 

A su vez, podemos ver que existe un ruido blanco en ambos espectrogramas, ya que este es homogéneo a través del tiempo en ambos, siendo menor en el celular que en la laptop. Por último, es importante recalcar que probablemente el ruido captado por el micrófono de la laptop sea mayoritariamente producido por los abanicos de la computadora, esto si bien no nos permite calcular exactamente el SNR del microfono en terminos practicos, nos da estimaciones exactas ya que al estar immplementados en una laptop es probable que siempre esten expuestos mínimamente a este ruido. 
Ahora bien, conociendo el SNR tenemos una idea clara de la calidad de los micrófonos y de como estos se desempeñan, sin embargo, procederemos a observar su respuesta en frecuencia.
